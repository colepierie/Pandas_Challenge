```python
import os
import pandas as pd
import csv
```

```python
# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data.head()
```

```python
Total_Players = len(purchase_data["SN"].unique())
Summary_df = pd.DataFrame({"Total Players": [Total_Players]})
```

```python
Summary_df
```

```python
unique_items = len(purchase_data["Item ID"].unique())
avg_price = purchase_data["Price"].mean()
total_purchases = purchase_data["Purchase ID"].count()
tot_rev = purchase_data["Price"].sum()
```

```python
summary2_df = pd.DataFrame({"number of unique items":[unique_items],
                            "Average Price": avg_price,
                            "Number of Purchases": total_purchases,
                             "Total Revenue": tot_rev})
summary2_df
```

```python
player_info = purchase_data.loc[:,["SN", "Age", "Gender"]].drop_duplicates()
```

```python
gender_data = player_info["Gender"].value_counts()
percent100 = player_info["Gender"].value_counts(normalize=True).mul(100).round(1).astype(str) + '%'
pd.DataFrame({'Total Count': gender_data, 'Percentage Of Players': percent100})

#consolidate same players
```

```python

avg_purch_price = purchase_data.groupby("Gender").Price.mean()
purch_count = purchase_data.groupby("Gender").count()["SN"]
total_purch_price = purchase_data.groupby("Gender").Price.sum()
purch_per_person = total_purch_price/gender_data


pd.DataFrame({"Purchase Count" : purch_count, "Average Purchase Price" : avg_purch_price,
              "Total Purchase Price" : total_purch_price, 
            "Avg Total Purchase Per Person" : purch_per_person})
```

```python
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-40", "40+"]
```

```python
#bins on age demo
player_info["Age Ranges"] = pd.cut(player_info["Age"], bins, labels=group_names, include_lowest=True)
player_info
```

```python
age_bin_count = player_info.groupby("Age Ranges").count()["SN"]
percent_age = age_bin_count/Total_Players

percent_age= percent_age.mul(100).round(1).astype(str) + '%'

pd.DataFrame({"Total Count": age_bin_count,
               "Percentage Of Players": percent_age}) 
```

```python
purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], bins, labels=group_names, include_lowest=True)
purchase_data
```

```python
age_group = purchase_data.groupby("Age Ranges").count()["Purchase ID"]
age_mean = purchase_data.groupby("Age Ranges").mean()["Price"].astype(float).map("${:,.2f}".format)
age_sum = purchase_data.groupby("Age Ranges").sum()["Price"].astype(float).map("${:,.2f}".format)
age_per = age_group/age_bin_count
age_per = age_per.astype(float).map("${:,.2f}".format)
#age_group.max()

#age_data = purchase_data["Age"]
```

```python
#Making DataFrame
pd.DataFrame({"Purchase Count": age_group,
              "Average Purchase Price": age_mean,
              "Total Purchase Value": age_sum,
              "Avg Total Purchase per Person": age_per})
```

```python
#groupby on SN
cust_group = purchase_data.groupby("SN").count()["Price"]
cust_mean = purchase_data.groupby("SN").mean()["Price"].astype(float).map("${:,.2f}".format)
cust_sum = purchase_data.groupby("SN").sum()["Price"].astype(float).map("${:,.2f}".format)

pd.DataFrame({"Purchase Count": cust_group,
              "Average Purchase Price": cust_mean,
              "Total Purchase Value": cust_sum}).sort_values(by="Total Purchase Value", ascending=False).head()


#find sum & mean & count
#sort to find top spenders decending order
```

```python
#create new dataframe with item info
#groupby for item name and id 2 collumns 
#sum mean price avg
#item_info = item_info.groupby(["Item ID", "Item Name"])

item_group = purchase_data.groupby(["Item ID", "Item Name"]).count()["Price"]
item_mean = purchase_data.groupby(["Item ID", "Item Name"]).mean()["Price"].astype(float).map("${:,.2f}".format)
item_sum = purchase_data.groupby(["Item ID", "Item Name"]).sum()["Price"].astype(float).map("${:,.2f}".format)

pd.DataFrame({"Purchase Count": item_group,
              "Item Price": item_mean,
              "Total Purchase Value": item_sum}).sort_values(by="Purchase Count", ascending=False).head()
```

```python
#sort dataframe by purchase total
pd.DataFrame({"Purchase Count": item_group,
              "Item Price": item_mean,
              "Total Purchase Value": item_sum}).sort_values(by="Total Purchase Value", ascending=False).head()


```
